پیدا کردن عنوان مقاله با کمک id

```python
r = Entrez.esummary(db = 'pubmed', id = '34030624')
rec = Entrez.parse(r)
for i in rec:
    print(i['Title'])
rec.close()
```

```python
h = Entrez.esummary( db="pubmed",  id='36326697' )
records = Entrez.parse(h)
for r  in  records:
    print( r['Title'] )
    print(r)
h.close() 
```
