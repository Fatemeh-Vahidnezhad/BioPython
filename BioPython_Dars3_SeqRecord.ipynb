# Write and read a file with fasta format

```python
from Bio.Seq import Seq
from Bio import SeqIO
from Bio.SeqRecord import SeqRecord
```

```python
myseq = Seq('MALWMRLLPLLALLALWGPDPAAAFVNQHLCGSHLVEALYLVCGERGFFYTPKTRREAEDLQVGQVELGGGPGAGSLQPLALEGSLQKRGIVEQCCTSICSLYQLENYCN')
str = SeqRecord(myseq, id = "1",description = 'INS')
str
```

```python
SeqIO.write(str,'str.fasta','fasta')
```

```python
file = SeqIO.read('INS.fasta',format = 'fasta')
file
```

```python
print(file.format('fasta'))
```
