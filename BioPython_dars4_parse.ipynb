```python
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord   #برای ثبت یک رکورد 
from Bio import SeqIO    #write and read a file
```

```python
seq1 = SeqIO.read('APP.fasta','fasta') #فقط یک توالی را می خواند و اگر دو توالی در یا فایل باشدُ خطا می دهد.
seq2 = SeqIO.read("INS.fasta","fasta")
```

```python
lst = [seq1, seq2]
```

```python
print(lst)
```

```python
SeqIO.write(lst, "myfile.fasta", "fasta") 
```

```python
r = SeqIO.parse('myfile.fasta',"fasta")#وقتی بیشتر از یک توالی در یک فایل باشد باید از این روش استفاده کرد
for i in r:
  print("\n",i.seq)
  print("\n", i.id)
  print("\n", len(i))
```

```python
r = SeqIO.parse('myfile.fasta',"fasta")#وقتی بیشتر از یک توالی در یک فایل باشد باید از این روش استفاده کرد
r1 = next(r) #no need loop function, using next for every seq
print(r1.seq)
r2 = next(r)
print(r2.seq)
```

```python
r = SeqIO.parse('myfile.fasta',"fasta")
r1 = next(r)
r1.id = "new_id"    #changing id
print(r1.format("fasta"))
```

```python
D = []
r = SeqIO.parse('myfile.fasta',"fasta")
r1 = next(r)
D.append(r1.seq)
r2 = next(r)
D.append(r2.seq)
print(D)
```

```python
D = []
S = []
all = []
r = SeqIO.parse('myfile.fasta',"fasta")
for i in r:
  D.append(i.description.split()[0])
  S.append(i.description.split()[1])
  all.append(i.description.split())

print(D)
print(S)
print(all)
```
